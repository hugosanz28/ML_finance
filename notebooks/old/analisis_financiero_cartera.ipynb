Vemos de cuales podemos obtener información a través de la API de yahoo finance y probamos con otras bolsas internacionales

```python
import yfinance as yf
import pandas as pd

# Sufijos de bolsas internacionales por si falla el de EE.UU.
ALT_SUFFIXES = [".BA", ".JK", ".VN", ".SN", ".CL", ".JO", ".KS", ".SA", ".BD"]

def is_valid(symbol: str, min_points: int = 3) -> bool:
    """Comprueba si Yahoo devuelve al menos algunos precios recientes."""
    try:
        df = yf.Ticker(symbol).history(period="10d", interval="1d")
        if df is not None and len(df.dropna()) >= min_points:
            # print(f"Simbolo válido: {symbol}.")
            return True
    except Exception:
        return False

def candidate_symbols(symbol: str):
    """Genera candidatos: original y variantes de sufijo."""
    out = []
    out.append(symbol)
    # Probar cambiar sufijo si tiene '.' y parece raro
    if "." in symbol:
        base = symbol.split(".")[0]
        for suf in ALT_SUFFIXES:
            out.append(base + suf)
    else:
        # si no tiene sufijo, probamos algunos
        for suf in ALT_SUFFIXES:
            out.append(symbol + suf)

    # deduplicar manteniendo orden
    seen = set()
    uniq = []
    for s in out:
        if s and s not in seen:
            seen.add(s)
            uniq.append(s)
    return uniq

def resolve_symbol(symbol: str):
    """Devuelve el primer candidato que tenga datos reales."""
    for cand in candidate_symbols(symbol):
        if is_valid(cand):
            return cand

# Resolvemos todos los tickers y rehacemos sectors ---
ticker_map = {}
report_rows = []

for sector, lst in sectors_union.items():
    for t in lst:
        if t in ticker_map:
            continue
        resolved = resolve_symbol(t)
        ticker_map[t] = resolved
        report_rows.append({
            "Sector": sector,
            "Original": t,
            "Resolved": resolved,
            "OK": resolved is not None
        })

report = pd.DataFrame(report_rows)

sectors_yf = {}
for sector, lst in sectors_union.items():
    resolved_list = [ticker_map[t] for t in lst if ticker_map.get(t)]
    # deduplicar
    sectors_yf[sector] = sorted(set(resolved_list))

print("Originales:", sum(len(v) for v in sectors_union.values()))
print("Resueltos:", sum(len(v) for v in sectors_yf.values()))
print("Fallidos:", (report["OK"] == False).sum())

# Ver fallidos
report[report["OK"] == False].sort_values(["Sector","Original"]).head(50)
```

Descargamos la información

```python
import numpy as np
from tqdm import tqdm
import yfinance as yf
import pandas as pd

all_tickers = sorted({t for sec in sectors_yf.values() for t in sec})

def download_in_batches(tickers, batch_size=25, **kwargs):
    frames = []
    for i in tqdm(range(0, len(tickers), batch_size)):
        batch = tickers[i:i+batch_size]
        df = yf.download(batch, **kwargs)
        frames.append(df)

    out = pd.concat(frames, axis=1)
    out = out.loc[:, ~out.columns.duplicated()]
    return out

data = download_in_batches(
    all_tickers,
    batch_size=25,
    start="2019-01-01",
    end="2025-12-01",
    interval="1d",
    auto_adjust=True,
    progress=False,
    threads=True
)

prices_raw = data["Close"]
```

```python
def build_sector_map_final(sector_map: dict, available_cols: set):
    sector_final = {}
    used_by_sector = {}
    missing_by_sector = {}

    for sector, tickers in sector_map.items():
        # quitamos duplicados manteniendo orden
        tickers_unique = list(dict.fromkeys(tickers))
        present = [t for t in tickers_unique if t in available_cols]
        missing = [t for t in tickers_unique if t not in available_cols]

        used_by_sector[sector] = present
        missing_by_sector[sector] = missing
        sector_final[sector] = present  # aquí ya solo los presentes

    return sector_final, used_by_sector, missing_by_sector

available = set(returns_used.columns)
sector_to_tickers_final, used, missing = build_sector_map_final(sectors_yf, available)

# Resúmenes
summary_used = pd.Series({sec: len(cols) for sec, cols in used.items()}).sort_values(ascending=False)
summary_missing = pd.Series({sec: len(cols) for sec, cols in missing.items() if len(cols) > 0}).sort_values(ascending=False)

print("Activos usados por sector (nº tickers presentes en el dataset final):")
display(summary_used)

print("\nSectores con tickers faltantes (nº tickers del sector que NO están en el dataset final):")
display(summary_missing)
```

Hallamos las correlaciones por sectores y graficamos las matrices de correlación.

```python
def corr_matrices_by_sector(data: pd.DataFrame, sector_to_tickers: dict, min_assets: int = 2):
    corr_by_sector = {}
    used_by_sector = {}
    missing_by_sector = {}

    available = set(data.columns)

    for sector, tickers in sector_to_tickers.items():
        tickers_unique = list(dict.fromkeys(tickers))
        present = [t for t in tickers_unique if t in available]
        missing = [t for t in tickers_unique if t not in available]

        used_by_sector[sector] = present
        missing_by_sector[sector] = missing

        if len(present) >= min_assets:
            corr_by_sector[sector] = data[present].corr()

    return corr_by_sector, used_by_sector, missing_by_sector

corr_mats, used2, missing2 = corr_matrices_by_sector(
    returns_used, sector_to_tickers_final, min_assets=2
)

print(f"Sectores con matriz calculada: {len(corr_mats)}")
```

```python
def plot_corr_matrix(corr: pd.DataFrame, title: str):
    n = len(corr.columns)
    figsize = (max(6, 0.55 * n), max(5, 0.55 * n))
    fig, ax = plt.subplots(figsize=figsize)
    im = ax.imshow(corr.values, aspect="auto")
    ax.set_xticks(range(n))
    ax.set_yticks(range(n))
    ax.set_xticklabels(corr.columns, rotation=90)
    ax.set_yticklabels(corr.index)
    ax.set_title(title)
    fig.colorbar(im, ax=ax)
    plt.tight_layout()
    plt.show()

for sector, corr in corr_mats.items():
    plot_corr_matrix(corr, f"Correlación (rendimientos) - {sector}")
```

## PCA

Vamos a realizar un análisis de coordendas principales basándonos en la matriz de distancias

```python
available = set(returns_used.columns)

sector_to_present = {
    sector: [t for t in tickers if t in available]
    for sector, tickers in sectors_yf.items()
}

sector_returns = pd.DataFrame(index=returns_used.index)
for sector, tickers in sector_to_present.items():
    if len(tickers) >= 2:
        sector_returns[sector] = returns_used[tickers].mean(axis=1, skipna=True)

sector_corr = sector_returns.corr()
sector_corr.head()

# Distancia basada en correlación
D = np.sqrt(0.5 * (1 - sector_corr.values))
```

```python
def pcoa_from_distance(D, n_components=2):
    """
    PCoA (Classical MDS) desde una matriz de distancias D (nxn).
    Devuelve coords (n x n_components), eigenvalues y proporción explicada.
    """
    D = np.asarray(D, dtype=float)
    n = D.shape[0]

    # Matriz de centrado
    J = np.eye(n) - np.ones((n, n)) / n

    # Doble centrado: B = -1/2 * J * D^2 * J
    B = -0.5 * J @ (D ** 2) @ J

    # Descomposición espectral (B simétrica)
    eigvals, eigvecs = np.linalg.eigh(B)

    # Orden descendente
    idx = np.argsort(eigvals)[::-1]
    eigvals = eigvals[idx]
    eigvecs = eigvecs[:, idx]

    # Nos quedamos con autovalores positivos (pueden salir pequeños negativos por redondeo)
    pos = eigvals > 1e-12
    eigvals_pos = eigvals[pos]
    eigvecs_pos = eigvecs[:, pos]

    # Coordenadas: V * sqrt(lambda)
    coords = eigvecs_pos[:, :n_components] * np.sqrt(eigvals_pos[:n_components])

    # Varianza explicada (sobre autovalores positivos)
    explained = eigvals_pos / eigvals_pos.sum()

    return coords, eigvals, explained

coords2, eigvals, explained = pcoa_from_distance(D, n_components=2)

print("Varianza explicada por los 2 primeros ejes:",
      explained[:2].sum() if len(explained) >= 2 else explained.sum())
```

La varianza explicada por los 2 primeros ejes no es alta

```python
def mean_intra_inter_corr(corr: pd.DataFrame, comm_map: dict):
    tickers = corr.index
    comm = pd.Series({t: comm_map.get(t, -1) for t in tickers})
    C = corr.copy()

    # Quitamos diagonal
    np.fill_diagonal(C.values, np.nan)

    # Intra: pares con misma comunidad
    intra_vals = []
    inter_vals = []
    for i, ti in enumerate(tickers):
        for j in range(i+1, len(tickers)):
            tj = tickers[j]
            rho = C.iat[i, j]
            if np.isnan(rho):
                continue
            if comm[ti] == comm[tj]:
                intra_vals.append(rho)
            else:
                inter_vals.append(rho)

    return np.nanmean(intra_vals), np.nanmean(inter_vals), len(intra_vals), len(inter_vals)

intra_mean, inter_mean, n_intra, n_inter = mean_intra_inter_corr(company_corr.loc[community_df.index, community_df.index], community_map)
print(f"Correlación media intra-comunidad: {intra_mean:.4f} (n={n_intra})")
print(f"Correlación media inter-comunidades: {inter_mean:.4f} (n={n_inter})")
```

## Algoritmo Genético

### Optimización in-sample

En primer lugar, usaremos todos los datos para hallar la cartera más eficiente y analizar si cumple nuestros supuestos.

Cargamos los datos, definimos retornos semanales, score por cartera, descendencia de los cruces de genes y mutaciones.

```python
import numpy as np
import pandas as pd
import math
#  1) Cargar datos 
path = "financial_data_clean.xlsx"   # ajusta ruta si hace falta
df = pd.read_excel(path, sheet_name="Returns", engine="openpyxl")
df["Date"] = pd.to_datetime(df["Date"])
df = df.set_index("Date").sort_index()
# Rendimientos diarios simples -> semanales simples
weekly = (1.0 + df).resample("W-FRI").prod() - 1.0
tickers = weekly.columns.to_list()
n = len(tickers)
mu_w = weekly.mean().to_numpy()
std_w = weekly.std(ddof=1).to_numpy()
corr_w = weekly.corr().to_numpy()
cov_w = np.outer(std_w, std_w) * corr_w  # cov semanal
# 2) GA: seleccionar 30 empresas con pesos iguales minimizando varianza 
def ga_min_var_equal_weight(
    tickers,
    mu_w,
    cov_w,
    k=20,
    target_return_annual=0.2,
    pop_size=260,
    generations=350,
    elite=20,
    tournament_size=5,
    crossover_rate=0.9,
    mutation_rate=0.25,
    penalty_scale=1e6,
    seed=123
):
    rng = np.random.default_rng(seed)
    n = len(tickers)
    target_w = target_return_annual / 52.0

    def random_ind():
        return np.sort(rng.choice(n, size=k, replace=False))

    def var_ret(ind):
        sub = cov_w[np.ix_(ind, ind)]
        var_w = float(sub.sum() / (k * k))
        ret_w = float(mu_w[ind].mean())
        return var_w, ret_w

    def fitness(ind):
        var_w, ret_w = var_ret(ind)
        short = max(0.0, target_w - ret_w)
        return var_w + penalty_scale * (short ** 2)

    pop = [random_ind() for _ in range(pop_size)]
    best = None
    best_fit = float("inf")

    for _ in range(generations):
        fits = np.array([fitness(ind) for ind in pop])
        order = np.argsort(fits)

        if fits[order[0]] < best_fit:
            best_fit = fits[order[0]]
            best = pop[order[0]].copy()

        new_pop = [pop[i].copy() for i in order[:elite]]

        def tournament():
            cand = rng.choice(pop_size, size=tournament_size, replace=False)
            return pop[cand[np.argmin(fits[cand])]]

        while len(new_pop) < pop_size:
            p1, p2 = tournament(), tournament()

            if rng.random() < crossover_rate:
                u = np.unique(np.concatenate([p1, p2]))
                if len(u) >= k:
                    c1 = np.sort(rng.choice(u, size=k, replace=False))
                    c2 = np.sort(rng.choice(u, size=k, replace=False))
                else:
                    rest = np.setdiff1d(np.arange(n), u)
                    c1 = np.sort(np.concatenate([u, rng.choice(rest, size=k - len(u), replace=False)]))
                    c2 = np.sort(np.concatenate([u, rng.choice(rest, size=k - len(u), replace=False)]))
            else:
                c1, c2 = p1.copy(), p2.copy()

            def mutate(ind):
                if rng.random() > mutation_rate:
                    return ind
                ind = ind.copy()
                pos = rng.integers(0, k)
                not_sel = np.setdiff1d(np.arange(n), ind)
                ind[pos] = rng.choice(not_sel)
                return np.sort(ind)

            new_pop.append(mutate(c1))
            if len(new_pop) < pop_size:
                new_pop.append(mutate(c2))

        pop = new_pop

    var_w, ret_w = var_ret(best)
    return (
        [tickers[i] for i in best],
        var_w * 52.0,
        ret_w * 52.0
    )

# ========= 3) Ejecutar =========
selected, var_annual, ret_annual = ga_min_var_equal_weight(
    tickers=tickers,
    mu_w=mu_w,
    cov_w=cov_w,
    k=30,
    target_return_annual=0.3 # Minimo retorno anual esperado
)
print("Cartera (30):", selected)
print(f"Varianza anual: {var_annual:.6f}")
print(f"Retorno anual esperado: {ret_annual*100:.2f}%")
print(f"Volatilidad anual: {math.sqrt(var_annual)*100:.2f}%")
```

```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt

path = "financial_data_clean.xlsx"   # ajusta ruta si hace falta
df = pd.read_excel(path, sheet_name="Returns", engine="openpyxl")
df["Date"] = pd.to_datetime(df["Date"])
df = df.set_index("Date").sort_index()
# Rendimientos diarios simples -> semanales simples
weekly = (1.0 + df).resample("W-FRI").prod() - 1.0
tickers = weekly.columns.to_list()
n = len(tickers)
mu_w = weekly.mean().to_numpy()
std_w = weekly.std(ddof=1).to_numpy()
corr_w = weekly.corr().to_numpy()
cov_w = np.outer(std_w, std_w) * corr_w  # cov semanal
# 2) GA: seleccionar 30 empresas con pesos iguales minimizando varianza 
def ga_min_var_equal_weight(
    tickers,
    mu_w,
    cov_w,
    k=20,
    target_return_annual=0.2,
    pop_size=260,
    generations=350,
    elite=20,
    tournament_size=5,
    crossover_rate=0.9,
    mutation_rate=0.25,
    penalty_scale=1e6,
    seed=123
):
    rng = np.random.default_rng(seed)
    n = len(tickers)
    target_w = target_return_annual / 52.0

    def random_ind():
        return np.sort(rng.choice(n, size=k, replace=False))

    def var_ret(ind):
        sub = cov_w[np.ix_(ind, ind)]
        var_w = float(sub.sum() / (k * k))
        ret_w = float(mu_w[ind].mean())
        return var_w, ret_w

    def fitness(ind):
        var_w, ret_w = var_ret(ind)
        short = max(0.0, target_w - ret_w)
        return var_w + penalty_scale * (short ** 2)

    pop = [random_ind() for _ in range(pop_size)]
    best = None
    best_fit = float("inf")

    for _ in range(generations):
        fits = np.array([fitness(ind) for ind in pop])
        order = np.argsort(fits)

        if fits[order[0]] < best_fit:
            best_fit = fits[order[0]]
            best = pop[order[0]].copy()

        new_pop = [pop[i].copy() for i in order[:elite]]

        def tournament():
            cand = rng.choice(pop_size, size=tournament_size, replace=False)
            return pop[cand[np.argmin(fits[cand])]]

        while len(new_pop) < pop_size:
            p1, p2 = tournament(), tournament()

            if rng.random() < crossover_rate:
                u = np.unique(np.concatenate([p1, p2]))
                if len(u) >= k:
                    c1 = np.sort(rng.choice(u, size=k, replace=False))
                    c2 = np.sort(rng.choice(u, size=k, replace=False))
                else:
                    rest = np.setdiff1d(np.arange(n), u)
                    c1 = np.sort(np.concatenate([u, rng.choice(rest, size=k - len(u), replace=False)]))
                    c2 = np.sort(np.concatenate([u, rng.choice(rest, size=k - len(u), replace=False)]))
            else:
                c1, c2 = p1.copy(), p2.copy()

            def mutate(ind):
                if rng.random() > mutation_rate:
                    return ind
                ind = ind.copy()
                pos = rng.integers(0, k)
                not_sel = np.setdiff1d(np.arange(n), ind)
                ind[pos] = rng.choice(not_sel)
                return np.sort(ind)

            new_pop.append(mutate(c1))
            if len(new_pop) < pop_size:
                new_pop.append(mutate(c2))

        pop = new_pop

    var_w, ret_w = var_ret(best)
    return (
        [tickers[i] for i in best],
        var_w * 52.0,
        ret_w * 52.0
    )


k = 25
targets = [x / 100 for x in range(14, 48, 2)]  # 14%, 16%, ..., 50%

vols = []   # volatilidad anual (%)
rets = []   # retorno anual obtenido (%)



for target_return in targets:
    selected, var_annual, ret_annual = ga_min_var_equal_weight(
        tickers=tickers,
        mu_w=mu_w,
        cov_w=cov_w,
        k=k,
        target_return_annual=target_return
    )

    vol_annual = math.sqrt(var_annual) * 100
    vols.append(vol_annual)
    rets.append(ret_annual * 100)

    print(
        f"Rentabilidad objetivo: {target_return*100:5.1f}% | "
        f"Volatilidad anual: {vol_annual:6.2f}%"
    )

# --- Gráfica: objetivo vs volatilidad ---
plt.figure(figsize=(9, 5))
plt.plot(vols,[t*100 for t in targets], marker="o")
plt.ylabel("Rentabilidad objetivo anual (%)")
plt.xlabel("Volatilidad anual (%)")
plt.title(f"Frontera (GA) | k={k} empresas | pesos iguales")
plt.grid(True)
plt.tight_layout()
plt.yscale("log")
plt.show()

# (opcional) también puedes graficar retorno REAL obtenido vs volatilidad
plt.figure(figsize=(9, 5))
plt.plot(vols, rets, marker="o")
plt.ylabel("Rentabilidad anual obtenida (%)")
plt.xlabel("Volatilidad anual (%)")
plt.title(f"Frontera (GA) real | k={k} empresas | pesos iguales")
plt.grid(True)
plt.tight_layout()
plt.yscale("log")
plt.show()
```

# Train y test

Por último vamos a formar una cartera con datos de hasta 2024 y ver cómo le hubiera ido en 2025

```python
import numpy as np
import pandas as pd
import math

# ========= 1) Cargar datos =========
path = "financial_data_clean.xlsx"
df = pd.read_excel(path, sheet_name="Returns", engine="openpyxl")
df["Date"] = pd.to_datetime(df["Date"])
df = df.set_index("Date").sort_index()

# ========= 2) Diario simple -> Semanal simple (composición) =========
weekly = (1.0 + df).resample("W-FRI").prod() - 1.0

# ========= 3) Split temporal =========
train_weekly = weekly.loc[: "2024-12-31"].copy()
test_weekly  = weekly.loc["2025-01-01" : "2025-12-31"].copy()

print("Train semanas:", train_weekly.shape[0], "| Test semanas:", test_weekly.shape[0])
```

```python
tickers = train_weekly.columns.to_list()
n = len(tickers)

mu_w = train_weekly.mean().to_numpy()
std_w = train_weekly.std(ddof=1).to_numpy()
corr_w = train_weekly.corr().to_numpy()
cov_w = np.outer(std_w, std_w) * corr_w
```

```python
def ga_min_var_equal_weight(
    tickers,
    mu_w,
    cov_w,
    k=20,
    target_return_annual=0.2,
    pop_size=260,
    generations=350,
    elite=20,
    tournament_size=5,
    crossover_rate=0.9,
    mutation_rate=0.25,
    penalty_scale=1e6,
    seed=123
):
    rng = np.random.default_rng(seed)
    n = len(tickers)
    target_w = target_return_annual / 52.0

    def random_ind():
        return np.sort(rng.choice(n, size=k, replace=False))

    def var_ret(ind):
        sub = cov_w[np.ix_(ind, ind)]
        var_w = float(sub.sum() / (k * k))
        ret_w = float(mu_w[ind].mean())
        return var_w, ret_w

    def fitness(ind):
        var_w, ret_w = var_ret(ind)
        short = max(0.0, target_w - ret_w)
        return var_w + penalty_scale * (short ** 2)

    pop = [random_ind() for _ in range(pop_size)]
    best = None
    best_fit = float("inf")

    for _ in range(generations):
        fits = np.array([fitness(ind) for ind in pop])
        order = np.argsort(fits)

        if fits[order[0]] < best_fit:
            best_fit = fits[order[0]]
            best = pop[order[0]].copy()

        new_pop = [pop[i].copy() for i in order[:elite]]

        def tournament():
            cand = rng.choice(pop_size, size=tournament_size, replace=False)
            return pop[cand[np.argmin(fits[cand])]]

        while len(new_pop) < pop_size:
            p1, p2 = tournament(), tournament()

            if rng.random() < crossover_rate:
                u = np.unique(np.concatenate([p1, p2]))
                if len(u) >= k:
                    c1 = np.sort(rng.choice(u, size=k, replace=False))
                    c2 = np.sort(rng.choice(u, size=k, replace=False))
                else:
                    rest = np.setdiff1d(np.arange(n), u)
                    c1 = np.sort(np.concatenate([u, rng.choice(rest, size=k - len(u), replace=False)]))
                    c2 = np.sort(np.concatenate([u, rng.choice(rest, size=k - len(u), replace=False)]))
            else:
                c1, c2 = p1.copy(), p2.copy()

            def mutate(ind):
                if rng.random() > mutation_rate:
                    return ind
                ind = ind.copy()
                pos = rng.integers(0, k)
                not_sel = np.setdiff1d(np.arange(n), ind)
                ind[pos] = rng.choice(not_sel)
                return np.sort(ind)

            new_pop.append(mutate(c1))
            if len(new_pop) < pop_size:
                new_pop.append(mutate(c2))

        pop = new_pop

    var_w, ret_w = var_ret(best)
    return (
        [tickers[i] for i in best],
        var_w * 52.0,
        ret_w * 52.0
    )
```

```python
# ===== Ejecutar GA (entrenamiento) =====
k_assets = 30
target = 0.12  # 12% anual objetivo

selected, var_w_train, ret_w_train = ga_min_var_equal_weight(
    tickers=tickers, mu_w=mu_w, cov_w=cov_w,
    k=k_assets,
    target_return_annual=target
)

print("Cartera seleccionada:", selected)
print(f"[TRAIN] Retorno anual aprox: {ret_w_train:.2%} | Vol anual aprox: {math.sqrt(var_w_train):.2%}")

# ===== Backtest en TEST (retorno realizado) =====
port_test_w = test_weekly[selected].mean(axis=1)  # equiponderada

# métricas realizadas (out-of-sample)
cum_return = (1 + port_test_w).prod() - 1
ann_return = (1 + cum_return) ** (52 / len(port_test_w)) - 1
ann_vol = port_test_w.std(ddof=1) * math.sqrt(52)

print("\n[TEST 2025] Resultados realizados (equiponderado):")
print(f"Retorno acumulado: {cum_return:.2%}")
print(f"Retorno anualizado: {ann_return:.2%}")
print(f"Volatilidad anualizada: {ann_vol:.2%}")
```
